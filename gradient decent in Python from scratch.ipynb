```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradient_descent(x,y):
    m_curr = b_curr =0  # we are starting for slope and gradient from 0.
    iterations = 1000   # No. of trials to get best slope and intercept for prediction equation
    n = len(x)
    learning_rate = 0.02  # cosidering a small unit which is going to impact the change for new slope and intercept.
    
    plt.figure(figsize=(15,10))
    plt.scatter(x,y, color='red',s =40, marker='o')  # Plot original x and y by using scatterplot
    
    for i in range(iterations):
        yp = m_curr * x + b_curr    # this is the equation for prediction
        
        cost = (1/n) * sum([val**2 for val in (y-yp)])   # this is loss(cost) fuction to evaluate loss in each itearation
        md = -(2/n) * sum(x*(y-yp))   # first derivatives of loss fuction w.r. to slope.
        bd = -(2/n) * sum(y-yp)       # first derivatives of loss fuction w.r. to intercept.
        
        m_curr = m_curr - learning_rate * md   # new slope
        b_curr = b_curr - learning_rate * bd   # new intercept
        print("m {}  , b {}  , cost {}  , iterations{}".format(m_curr, b_curr, cost, i))
        plt.plot(x,yp, color ='green', ls='solid')
    pass


```

```python
gradient_descent(x,y)
```
